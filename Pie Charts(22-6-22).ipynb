```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=['r','b','pink']
la=['DSP','CD','WT']
plt.pie(y,colors=co,labels=la)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=['r','b','pink']
la=['DSP','CD','WT']
plt.pie(y,colors=co,labels=la)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=['r','b','pink']
la=['DSP','CD','WT']
plt.pie(y,colors=co,labels=la)
plt.legend(title="Subjects")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=['r','b','pink']
la=['DSP','CD','WT']
myexplode=[0.2,0,0]
plt.pie(y,colors=co,labels=la,explode=myexplode)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
co=['r','b','pink']
la=['DSP','CD','WT']
myexplode=[0.2,0,0]
plt.pie(y,colors=co,labels=la,explode=myexplode,shadow=True)
plt.legend()
plt.show()
```
