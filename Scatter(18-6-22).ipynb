# Scatter

```python
import matplotlib.pyplot as plt
x=[5,8,4,2,6,9,7]
y=[10,50,30,20,40,70,60]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,3,5,7,11,13]
y=[1,2,4,6,8,9]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,3,5,7,11,13]
y=[1,2,4,6,8,9]
plt.scatter(x,y,c='red')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,3,5,7,11,13]
y=[1,2,4,6,8,9]
plt.scatter(x,y,c='green',marker='^')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,3,5,7,11,13]
y=[1,2,4,6,8,9]
plt.scatter(x,y,c='red',marker='>')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,3,5,7,11,13]
y=[1,2,4,6,8,9]
colors=['red','green','pink','blue','yellow','black']
plt.scatter(x,y,c=colors)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,3,5,7,11,13]
y=[1,2,4,6,8,9]
plt.scatter(x,y,c='red',alpha=0.9)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,3,5,7,11,13]
y=[1,2,4,6,8,9]
plt.scatter(x,y,marker='*')
plt.show()
```
