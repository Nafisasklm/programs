```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([1,5])
ypoints=np.array([10,50])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot([4,60],[7,150],"--")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y)
plt.show
```

```python
#Multiple points
xpoints=np.array([2,3,5,7])
ypoints=np.array([6,4,2,3])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#plotting withour x-points
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints)
plt.show()
```

```python
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints,marker="*",mec="r")
plt.show()
```

```python
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints,marker="o",mec="b",color="r")
plt.show()
```

```python
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints,marker="*",mec="r",linewidth="10")
plt.show()
```
